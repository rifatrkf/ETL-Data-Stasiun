```python
import pandas as pd

data = pd.read_csv("D:\Magang\Wikimedia\quick Statements\Sisa\data stasiun no 17 - akhir.csv")
print(data.head())
```

Menghapus kolom wilayah

```python
data.isnull().sum()
```

```python
data_bersih = data.dropna()
```

```python
data_bersih.describe()
```

```python
data_bersih["Koordinat"] = data_bersih["Koordinat"].replace(", ", "/", regex=True)
```

```python
data_bersih["Koordinat"] = "@" + data_bersih["Koordinat"]
```

```python
data_bersih["daerah administratif"] = data_bersih["daerah administratif"].replace("Maritim ", "", regex=True)
```

```python
data_bersih["daerah administratif"] = data_bersih["daerah administratif"].replace("Dok II ", "", regex = True)
```

```python
data_bersih
```

```python
jenis_stasiun = []
a = list(data_bersih["Stasiun / UPT"])
for i, x in enumerate(a):
    if "Meteorologi" in a[i]:
        jstasiun = "Stasiun Meteorologi"
    elif "Klimatologi" in a[i]:
        jstasiun = "Stasiun Klimatologi"
    elif "Geofisika" in a[i]:
        jstasiun = "Stasiun Geofisika"
    jenis_stasiun.append(jstasiun)
jenis_stasiun
```

```python
kelas = []
for i, x in enumerate(a):
    if "Kelas IV" in a[i]:
        jkelas = "Kelas IV"
    elif "Kelas III" in a[i]:
        jkelas = "Kelas III"
    elif "Kelas II" in a[i]:
        jkelas = "Kelas II"
    elif "Kelas I" in a[i]:
        jkelas = "Kelas I"
    kelas.append(jkelas)
kelas
```

```python
jenis_wilayah = list(data_bersih["Wilayah"])
nama_tempat = list(data_bersih["daerah administratif"])
print(jenis_wilayah)
print(nama_tempat)
```

```python
deskripsi = []
for i, x in enumerate(a):
    teks ="{} {} Balai Besar{} yang terletak di {}".format(jenis_stasiun[i], kelas[i], jenis_wilayah[i], nama_tempat[i])
    deskripsi.append(teks)
deskripsi
```

```python
data_bersih.insert(1, "Deskripsi", deskripsi, True)
```
